```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report 
```

```python
#loading the dataset
data = pd.read_csv(r'E:\Users\Nirvan\Desktop\Projects\spam.csv' , encoding='latin1')
```

```python
print(data.shape)
```

```python
data.head()
```

```python
data = data.rename(columns = {"v1":"label", "v2":"email_text"})
```

```python
data = data.drop(columns = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"])
```

```python
print(data['label'].value_counts())
```

```python
print(data.isnull().sum())
```

```python
data['label'] = data['label'].map({'ham':0, 'spam':1, 'Spam':1})
```

```python
data.head()
```

```python
data = data[["email_text", "label"]]
```

```python
data.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data['email_text'], data['label'], test_size=0.2, random_state=0)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vectorized, Y_train)
```

```python
Y_pred = model.predict(X_test_vectorized)
print("accuracy: ", accuracy_score(Y_test, Y_pred))
print("classification report: ", classification_report(Y_test, Y_pred))
```

```python
test_email = ["congratulations! You won a free holiday!"]
test_vectorized = vectorizer.transform(test_email)
prediction = model.predict(test_vectorized)
print("Prediction: ", "Spam" if prediction[0] == 1 else "Not spam")
```
