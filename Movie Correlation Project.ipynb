```python
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) 

# Importing the data

df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype('int64')
```

```python
# Data types
df.dtypes
```

```python
#Changing data types
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#High correlations
```

```python
#Scatter plot with budget vs. gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs. Gross')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget')

plt.show()
```

```python
# Plot budget vs. gross using seaborne

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Correlation

df.corr(method='spearman') #pearson, kendall, spearman
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_matrix = df.corr(method='spearman')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix (Spearman method)')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
corr_matrix = df_numerized.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
